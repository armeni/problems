# Плотность чисел с плавающей запятой

```python
# расстояние до ближайшего числа, большего x
def eps(x):
    d = x / 2

    x1 = x + d
    x1_prev = x

    while True:
        if x1 == x1_prev:
            return d
        
        d /= 2
        x1_prev = x1
        x1 = x + d
```

Смотрим, как меняется расстояние между числами float32 в зависимости от их величины (пропорционально)

```python
plt.figure()

xs = 10. ** np.arange(-10, 11)
ys = [eps(x) for x in xs]
plt.plot(np.log2(xs), np.log2(ys), 'k.-')

xs = 2. ** np.arange(-30, 31)
ys = [eps(x) for x in xs]
plt.plot(np.log2(xs), np.log2(ys), 'm.')

plt.xlabel('log2(x)')
plt.ylabel('log2(eps(x))')
pass
```

Проверим ошибку округления для float32 (воспользуемся для этого float64)

```python
x0 = 1
x64 = np.linspace(-x0, x0, 1000, dtype='float64')
x32 = x64.astype(dtype='float32')

plt.figure()
plt.plot(x64, np.abs(x32-x64), 'k.-')
plt.xlabel('x')
plt.ylabel('float32 vs float64 error')
pass
```
